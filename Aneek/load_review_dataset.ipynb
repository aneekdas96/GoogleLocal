```python
import json
import pandas as pd 
import numpy as np 
import re
import yaml
import ast
```

```python
f = open('./review_data/reviews.json', encoding='utf-8')
```

```python
master_list = []
for i in range(100):
    line = f.readline()
    line = line.replace("u'", '"')
    line = line.replace("'", '"')
    try:
        d = ast.literal_eval(line)
        buff = [d[key] for key in d.keys()]
        if buff[1].isascii() == True and buff[2].isascii() == True:
            master_list.append(buff)
    except Exception as e:
        pass
```

```python
columns = [key for key in d.keys()]
master_df = pd.DataFrame(master_list, columns=columns)
```
